```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset('titanic')
```

```python
data
```

```python


data['age'] = data['age'].fillna(np.mean(data['age']))
data['embarked'] = data['embarked'].fillna(data['embarked'].mode()[0])
```

```python

data.isnull().sum()
```

```python
titanic = data.dropna(subset=['age', 'sex', 'survived'])

# Plot boxplot: Age distribution by Sex and Survival
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')

plt.title('Box Plot of Age Distribution by Gender and Survival')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=titanic, x='sex', y='age', hue='survived', split=True, palette='Set2')
plt.title('Violin Plot of Age Distribution by Gender and Survival')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(data=titanic, x='sex', y='age', hue='survived', dodge=True, palette='Set1')
plt.title('Swarm Plot of Age Distribution by Gender and Survival')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', kde=True, bins=30, palette='muted')
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.grid(True)
plt.show()
```
